```python
# Visual test for CaptureLiveCellstoCSV_sorted.lua output
# Reads classified_boundaries.csv and plots outer, inner, then both 

import pandas as pd
import plotly.graph_objects as go
from pathlib import Path

# Path to CSV
csv_path = Path("CaptureBugBoundary/classified_boundaries.csv")
if not csv_path.exists():
    csv_path = Path("../CaptureBugBoundary/classified_boundaries.csv")
df = pd.read_csv(csv_path)

outer = df[df["boundary_type"] == "outer"]
inner = df[df["boundary_type"] == "inner"]
print(f"Loaded: {len(outer)} outer, {len(inner)} inner boundary points")
```

## 1. Outer boundaries only
Points that touch the "external" background (outside the shape).

```python
fig_outer = go.Figure()
fig_outer.add_trace(go.Scatter(
    x=outer["x"], y=outer["y"],
    mode="markers", name="outer",
    marker=dict(size=4, color="steelblue", symbol="square")
))
fig_outer.update_layout(
    title="Outer boundaries only",
    xaxis_title="x", yaxis_title="y",
    xaxis=dict(scaleanchor="y", scaleratio=1),
    yaxis=dict(autorange="reversed"),
    height=500,
)
fig_outer.show()
```

## 2. Inner boundaries only
Points that touch holes (dead regions inside the shape).

```python
fig_inner = go.Figure()
fig_inner.add_trace(go.Scatter(
    x=inner["x"], y=inner["y"],
    mode="markers", name="inner",
    marker=dict(size=4, color="coral", symbol="circle")
))
fig_inner.update_layout(
    title="Inner boundaries only",
    xaxis_title="x", yaxis_title="y",
    xaxis=dict(scaleanchor="y", scaleratio=1),
    yaxis=dict(autorange="reversed"),
    height=500,
)
fig_inner.show()
```

## 3. Both on same plot 
Outer and inner boundaries together

```python
fig_both = go.Figure()
fig_both.add_trace(go.Scatter(
    x=outer["x"], y=outer["y"],
    mode="markers", name="outer",
    marker=dict(size=4, color="steelblue", symbol="square", opacity=0.8)
))
fig_both.add_trace(go.Scatter(
    x=inner["x"], y=inner["y"],
    mode="markers", name="inner",
    marker=dict(size=4, color="coral", symbol="circle", opacity=0.8)
))
fig_both.update_layout(
    title="Outer + inner boundaries (zoom to inspect overlap)",
    xaxis_title="x", yaxis_title="y",
    xaxis=dict(scaleanchor="y", scaleratio=1),
    yaxis=dict(autorange="reversed"),
    height=600,
    legend=dict(yanchor="top", y=0.99, xanchor="left", x=0.01),
)
fig_both.show()
```
